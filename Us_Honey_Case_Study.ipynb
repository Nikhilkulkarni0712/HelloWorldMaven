```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/US_honey_dataset (1).csv')
```

```python
df
```

```python
plt.figsize(20,8)
```

```python
#top and bottom 5 states vs Avg_price,production,value of production
```

```python
States_Production=df.groupby('state')['production'].sum().reset_index()
```

```python
States_Production
```

```python
States_Production.head()
```

```python
#write the code for top 5 states producing honey
```

```python
States_Production.sort_values("production", ascending=False).head()
```

```python
sns.barplot(x='state',y='production',data=States_Production.sort_values("production", ascending=False).head())
plt.title("Top 5 states producing honey")
plt.xlabel("States")
plt.ylabel("Production")
plt.show()
```

```python
#give me the 5 worst performing states based upon production
```

```python
sns.barplot(x='state',y='production',data=States_Production.sort_values("production", ascending=True).head())
plt.title("Top 5 worst performing states producing honey")
plt.xlabel("States")
plt.ylabel("Production")
plt.show()
```

```python
States_Production.sort_values("production", ascending=True).head()
```

```python
States_Avg_price=df.groupby('state')['average_price'].sum().reset_index()
States_Avg_price
```

```python
States_Avg_price.sort_values("average_price", ascending=True).head()
```

```python
sns.barplot(x='state',y='average_price',data=States_Avg_price.sort_values("average_price", ascending=True).head())
plt.title("Bottom 5 states producing honey based on avg price")
plt.xlabel("States")
plt.ylabel("Production")
plt.show()
```

```python
plt.pie(x = 'average_price',labels = 'state', autopct = '%0.2f%%',explode = [0,0,0,0,0.2],data = States_Avg_price.sort_values("average_price", ascending=True).head())
plt.title('bottom 5 states based in US')
plt.show()
```

```python
plt.figure(figsize=(20, 15))

# Top 5 states based on production
plt.subplot(2, 2, 1) # 2 rows, 2 columns, first subplot
sns.barplot(x='state', y='production', data=States_Production.sort_values("production", ascending=False).head())
plt.title('Top 5 States by Honey Production')
plt.xlabel('State')
plt.ylabel('Production')

# Bottom 5 states based on production
plt.subplot(2, 2, 2) # 2 rows, 2 columns, second subplot
sns.barplot(x='state', y='production', data=States_Production.sort_values("production", ascending=True).head())
plt.title('Bottom 5 States by Honey Production')
plt.xlabel('State')
plt.ylabel('Production')

# Top 5 states based on average price
plt.subplot(2, 2, 3) # 2 rows, 2 columns, third subplot
sns.barplot(x='state', y='average_price', data=States_Avg_price.sort_values("average_price", ascending=False).head())
plt.title('Top 5 States by Average Honey Price')
plt.xlabel('State')
plt.ylabel('Average Price')

# Bottom 5 states based on average price
plt.subplot(2, 2, 4) # 2 rows, 2 columns, fourth subplot
sns.barplot(x='state', y='average_price', data=States_Avg_price.sort_values("average_price", ascending=True).head())
plt.title('Bottom 5 States by Average Honey Price')
plt.xlabel('State')
plt.ylabel('Average Price')

#plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 15))

plt.subplot(2, 2, 1)
sns.barplot(x='state', y='production', data=States_Production.sort_values("production", ascending=False).head())
plt.title('Top 5 States by Honey Production')
plt.xlabel('State')
plt.ylabel('Production')

plt.subplot(2, 2, 2)
sns.barplot(x='state', y='production', data=States_Production.sort_values("production", ascending=True).head())
plt.title('Bottom 5 States by Honey Production')
plt.xlabel('State')
plt.ylabel('Production')

plt.subplot(2, 2, 3)
sns.barplot(x='state', y='average_price', data=States_Avg_price.sort_values("average_price", ascending=False).head())
plt.title('Top 5 States by Average Honey Price')
plt.xlabel('State')
plt.ylabel('Average Price')


plt.subplot(2, 2, 4)
sns.barplot(x='state', y='average_price', data=States_Avg_price.sort_values("average_price", ascending=True).head())
plt.title('Bottom 5 States by Average Honey Price')
plt.xlabel('State')
plt.ylabel('Average Price')

plt.show()
```
